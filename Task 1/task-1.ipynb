```python
# Rock-Paper-Scissors Game-Task 1

import random   # importing random module so the computer can make random choices

choices = ["rock", "paper", "scissors"]  # list of valid choices

while True:  # loop to keep the game running
    
    # asking user to choose an option
    user = input("Choose rock, paper, or scissors (or 'quit' to stop): ").lower()
    
    if user == "quit":   # if user types quit â†’ exit game
        print("Game Ended.")
        break
    
    if user not in choices:  # checking if user entered a valid choice
        print("Invalid choice. Try again.")
        continue
    
    ai = random.choice(choices)  # computer makes a random choice from the list
    print(f"Computer chose: {ai}")  # showing computer's choice
    
    # checking game rules to decide winner
    if user == ai:
        print("It's a tie!")
    elif (user == "rock" and ai == "scissors") or \
         (user == "paper" and ai == "rock") or \
         (user == "scissors" and ai == "paper"):
        print("You win!")
    else:
        print("You lose!")
```
